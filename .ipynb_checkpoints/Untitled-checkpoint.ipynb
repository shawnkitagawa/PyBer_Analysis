```python
import matplotlib.pyplot as plt
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
plt.show()
```

```python
plt.plot(x_axis, y_axis, label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
plt.figure(figsize = (20,20))
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax.plot(x_axis, y_axis, color = "green", marker = "D", linewidth = 2, label = "Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fare($)")
ax.set_xlabel("Date")
ax.set_ylim(0,45)
ax.grid()





```

```python
plt.bar(x_axis, y_axis, color = "green", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.title("Pyber Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")


plt.legend()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "black", label = "New York")
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")

plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s = y_axis)
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data* 3)
    
```

```python
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = y_axis, color = "skyblue", alpha = .2, edgecolor = "black", linewidth = 2, label = "Boston")
ax.gca().invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

ax.legend()
```
